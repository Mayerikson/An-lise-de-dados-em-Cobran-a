##1.1 Como Funciona a Área de Cobrança de um Banco.

A área de cobrança é responsável por recuperar valores devidos de clientes que estão inadimplentes ou com pagamentos atrasados. Seu objetivo principal é minimizar perdas financeiras e maximizar a recuperação de créditos e para isso ela utiliza o **Score Collections**.

O Score de Cobrança (ou Score Collections ) é uma ferramenta utilizada por instituições financeiras para avaliar a probabilidade de um cliente inadimplente quitar sua dívida dentro de um determinado período. Ele funciona como um sistema de pontuação que classifica os clientes com base em diversos fatores, como comportamento de pagamento, perfil socioeconômico, histórico de crédito e outras variáveis relevantes.



O objetivo principal do Score de Cobrança é priorizar esforços de recuperação de crédito , permitindo que a equipe de cobrança concentre seus recursos nos casos mais promissores ou críticos. Isso ajuda a otimizar o processo de cobrança, aumentando a eficiência e reduzindo custos operacionais.



##1.2 Análise SWOT da Área de Cobrança

A análise SWOT ajuda a identificar forças internas, fraquezas internas, oportunidades externas e ameaças externas. Aqui está uma análise detalhada:

Forças (Strengths)

Tecnologia Avançada: Bancos modernos possuem sistemas sofisticados de gestão de cobrança, incluindo plataformas de CRM (Customer Relationship Management) e IA (Inteligência Artificial).

Equipe Treinada: Profissionais capacitados para lidar com diferentes perfis de clientes e situações complexas.

Dados Abundantes: Grande volume de dados históricos sobre clientes, permitindo análises profundas e personalização das abordagens.

Processos Estruturados: Fluxos de trabalho bem definidos garantem consistência e eficiência nos processos de cobrança.

Fraquezas (Weaknesses)

Resistência ao Pagamento: Alguns clientes podem ser refratários a pagamentos, mesmo após várias tentativas de contato.

Custos Operacionais Altos: Manter uma equipe de cobrança pode ser caro, especialmente em grandes carteiras de crédito.

Desgaste com Clientes: Abordagens agressivas podem prejudicar a reputação do banco e afastar clientes.

Dependência de Tecnologia: Problemas técnicos ou falhas nos sistemas podem comprometer a eficácia da cobrança.

Oportunidades (Opportunities)

Automatização: Implementar chatbots e outras soluções de automação para reduzir custos e aumentar a escala.

Personalização: Usar machine learning para oferecer soluções customizadas aos clientes, aumentando as chances de sucesso nas negociações.

Parcerias Comerciais: Colaborar com empresas de recuperação de crédito especializadas para casos difíceis.

Inovação em Produtos Financeiros: Criar produtos específicos para clientes com histórico de inadimplência, como linhas de crédito com condições especiais.

Ameaças (Threats)

Regulamentação: Mudanças nas leis e regulamentos podem impor restrições adicionais às práticas de cobrança.

Concorrência: Bancos rivais podem oferecer melhores condições de renegociação, atraindo clientes inadimplentes.

Crise Econômica: Periodos de recessão podem aumentar a inadimplência geral, tornando a recuperação de créditos mais difícil.

Reputação Negativa: Práticas agressivas de cobrança podem danificar a imagem do banco junto ao público.

##1.3 Análise SMART da Área de Cobrança

O framework SMART serve para definir objetivos claros e mensuráveis. Aqui estão exemplos de metas SMART para a área de cobrança:

Objetivo SMART 1: Reduzir o Tempo Médio de Recuperação de Créditos.

Específico (Specific): Reduzir o tempo médio de recuperação de créditos de 60 dias para 45 dias.

Medível (Measurable): Monitorar o tempo médio de recuperação usando métricas de KPIs (Indicadores-Chave de Desempenho).

Alcançável (Achievable): Implementar processos automáticos de notificação e modelos preditivos para identificar clientes com maior probabilidade de pagamento rápido.

Relevante (Relevant): Essa meta é crucial para melhorar a liquidez do banco e reduzir perdas financeiras.

Temporal (Time-bound): Alcançar esse objetivo em 6 meses.

Objetivo SMART 2: Aumentar a Taxa de Adesão a Acordos de Renegociação

Específico (Specific): Aumentar a taxa de adesão a acordos de renegociação de 30% para 45%.

Medível (Measurable): Registrar o número de acordos fechados versus o total de clientes inadimplentes.

Alcançável (Achievable): Personalizar ofertas de renegociação com base no perfil do cliente, utilizando análise de dados.

Relevante (Relevant): Isso ajudará a preservar relacionamentos com clientes e reduzir perdas significativamente.

Temporal (Time-bound): Alcançar essa meta em 1 ano.

Objetivo SMART 3: Reduzir o Número de Queixas de Clientes

Específico (Specific): Reduzir o número de queixas de clientes relacionadas à cobrança em 50%.

Medível (Measurable): Monitorar o número de reclamações recebidas via SAC

 (Serviço de Atendimento ao Consumidor).

Alcançável (Achievable): Treinar a equipe de cobrança para abordagens mais empáticas e transparentes, além de implementar gravações de chamadas para monitorar a qualidade das interações.

Relevante (Relevant): Reduzir queixas melhora a satisfação do cliente e protege a reputação do banco.

Temporal (Time-bound): Alcançar essa meta em 6 meses.

## 3.2  Profissão

```python
# Total

df["Profissão"].value_counts()
```

```python
# Contagem de valores na coluna "Profissão"
counts = df["Profissão"].value_counts()
total = len(df)  # Total de registros no DataFrame

# Criar o gráfico
plt.figure(figsize=(10, 6))  # Ajusta o tamanho do gráfico

# Usar uma paleta de cores com mais tons para evitar repetição
colors = sns.color_palette("husl", n_colors=len(counts))  # Paleta com número dinâmico de cores

ax = sns.barplot(
    x=counts.index,
    y=counts.values,
    hue=counts.index,  # Atribuir a variável 'x' ao parâmetro 'hue'
    palette=colors,    # Usar a paleta dinâmica
    legend=False       # Desativar a legenda
)

# Adicionar as porcentagens dentro das barras
for i, count in enumerate(counts.values):
    percentage = (count / total) * 100  # Calcula a porcentagem
    ax.text(
        i,  # Posição x (índice da barra)
        count + 0.02 * max(counts),  # Posição y (ligeiramente acima da barra)
        f"{percentage:.1f}%",  # Texto com a porcentagem formatada
        ha="center",  # Alinhamento horizontal
        fontsize=10,  # Tamanho da fonte
        color="black"  # Cor do texto
    )

# Personalizar o gráfico
plt.title("Distribuição de Profissões", fontsize=14, fontweight="bold")
plt.xticks(rotation=45, fontsize=10)  # Rotaciona os rótulos do eixo x
plt.yticks(fontsize=10)
plt.xlabel("")  # Remove o rótulo do eixo x
plt.ylabel("")  # Remove o rótulo do eixo y
sns.despine()  # Remove as bordas do gráfico

# Mostrar o gráfico
plt.tight_layout()  # Ajusta o layout para evitar sobreposições
plt.show()
```

Insights:

O gráfico de distribuição de profissões fornece uma visão clara da composição ocupacional dos clientes ou indivíduos no conjunto de dados. As profissões mais comuns (Gerente de Mídia e Advogado) destacam-se, enquanto outras, como Escritor, têm menor representatividade

# 4.0 Análise Bivariada`


As análises bidimensionais são úteis para identificar possíveis relações entre duas variáveis. Elas ajudam a entender como diferentes variáveis interagem e como essas interações podem influenciar Pagou a dívida?.


Para que possamos identificar quais fatores possuem uma forte relação com o Pagou a dívida? e selecionar os dados que serão preparados, vou realizar uma análise bidimensional através do Information Value (IV).


Cálculo do Information Value


O cálculo do Information Value (IV), uma técnica estatística utilizada para avaliar o poder preditivo de variáveis independentes em relação a uma variável de resultado.


Quanto maior o IV de uma variável, maior é a sua capacidade de discriminar entre as categorias da variável resultado.

Os valores de IV podem ser interpretados da seguinte forma:


image.png



```python

def binarizar_variavel(df, feature, bins=5):
    """Agrupa variáveis numéricas em faixas usando quantis, mantendo o nome original da variável."""
    if df[feature].dtype != 'object':
        df[feature] = pd.qcut(df[feature], q=bins, duplicates='drop')
    return feature

def calcular_woe_iv_prob(df, feature, target, woe_tables):
    try:
        df_clean = df[[feature, target]].dropna()
        total_good = df_clean[df_clean[target] == 1].shape[0]
        total_bad = df_clean[df_clean[target] == 0].shape[0]

        if total_good == 0 or total_bad == 0:
            return 0

        # Adiciona observed=False para evitar o FutureWarning
        agrupado = df_clean.groupby(feature, observed=False)[target].agg(['count', 'sum'])
        agrupado.columns = ['Total', 'Pagou']
        agrupado['Não_Pagou'] = agrupado['Total'] - agrupado['Pagou']

        agrupado['Distr_Good'] = agrupado['Pagou'] / total_good
        agrupado['Distr_Bad'] = agrupado['Não_Pagou'] / total_bad

        agrupado['WOE'] = np.log((agrupado['Distr_Good'] + 1e-10) / (agrupado['Distr_Bad'] + 1e-10))
        agrupado['IV'] = (agrupado['Distr_Good'] - agrupado['Distr_Bad']) * agrupado['WOE']

        agrupado['Prob_Pagou'] = (agrupado['Pagou'] / agrupado['Total'] * 100).round(2).astype(str) + '%'
        agrupado['Distr_Good'] = (agrupado['Distr_Good'] * 100).round(2).astype(str) + '%'
        agrupado['Distr_Bad'] = (agrupado['Distr_Bad'] * 100).round(2).astype(str) + '%'

        woe_tables[feature] = agrupado.reset_index()

        return agrupado['IV'].sum()
    except Exception as e:
        print(f"Erro ao processar {feature}: {e}")
        return 0

def classificar_iv(valor):
    if valor < 0.02: return 'Inútil'
    elif valor < 0.1: return 'Fraco'
    elif valor < 0.3: return 'Médio'
    elif valor < 0.5: return 'Forte'
    else: return 'Muito Forte'

# Execução principal
if __name__ == "__main__":
    df = pd.read_excel('/content/base_de_dados_cobranca.xlsx', sheet_name='Base')
    target = 'Pagou a dívida?'

    features = [
        'Profissão',
        'Renda Comprovada',
        'Dias em Atraso',
        'Qte_Atrasos_Histórico',
        'Saldo_Devedor',
        'Parcelas Restantes',
        'Emprestimo_com_Garantia?'  # Variável adicionada
    ]

    # Tratar nulos
    for col in features:
        if df[col].dtype != 'object':
            df[col] = df[col].fillna(df[col].median())
        else:
            df[col] = df[col].fillna(df[col].mode()[0])

    # Reduzir outliers (apenas para numéricas)
    for col in features:
        if df[col].dtype != 'object':
            q1 = df[col].quantile(0.25)
            q3 = df[col].quantile(0.75)
            iqr = q3 - q1
            df[col] = np.clip(df[col], q1 - 1.5 * iqr, q3 + 1.5 * iqr)

    woe_tables = {}
    iv_dict = {}

    # Agrupar e calcular métricas
    for col in features:
        agrupada = binarizar_variavel(df, col)
        iv = calcular_woe_iv_prob(df, agrupada, target, woe_tables)
        iv_dict[col] = iv

    # Tabela de IV total
    iv_df = pd.DataFrame.from_dict(iv_dict, orient='index', columns=['IV'])
    iv_df['Classificação'] = iv_df['IV'].apply(classificar_iv)
    iv_df = iv_df.sort_values('IV', ascending=False)

    print("\n" + "="*60)
    print("Análise de Poder Preditivo (Information Value)")
    print("="*60)
    print(iv_df.to_string())

    # Exibir tabelas WOE
    for feature, tabela in woe_tables.items():
        print("\n" + "-"*50)
        print(f"Tabela WOE - {feature}")
        print(tabela.to_string(index=False))

    # Gráfico de barras
    plt.figure(figsize=(10, 6))
    bars = plt.bar(iv_df.index, iv_df['IV'], color='teal')
    for bar in bars:
        plt.text(bar.get_x() + bar.get_width()/2., bar.get_height(),
                 f'{bar.get_height():.4f}', ha='center', va='bottom')

    for val, color, label in [(0.02, 'red', 'Inútil'), (0.1, 'orange', 'Fraco'),
                              (0.3, 'green', 'Médio'), (0.5, 'blue', 'Forte')]:
        plt.axhline(val, color=color, linestyle='--')
        plt.text(len(iv_df) - 0.5, val + 0.01, label, color=color)

    plt.title('Information Value por Variável')
    plt.xticks(rotation=45)
    plt.ylabel('IV')
    plt.tight_layout()
    plt.show()
```
